```python
import os, pickle
import numpy as np
import pandas as pd

curr_path = os.getcwd()
data_path = os.path.abspath(os.path.join(curr_path, '../data'))
```

```python
def convertDataFile(idx, path, df):
    try:
        with open(path, 'rb') as file:
            file_data = pickle.load(file)

        file_df = pd.DataFrame()

        for data in file_data:
            if data['valid'] and data['handData']:
                word = [data['word']]
                handData = np.array(data['handData'])

                if handData[:,0].shape[0] > 60:
                    run_index = np.repeat([idx], handData.shape[0]).reshape(handData.shape[0], 1)
                    word = np.repeat(word, handData.shape[0]).reshape(handData.shape[0], 1)

                    data_df = pd.DataFrame(np.hstack([run_index, word, handData]))
                    data_df.sort_values(by=17, inplace=True)
                    data_df.reset_index(drop=True)

                    file_df = pd.concat([file_df, data_df], ignore_index=True)

        df = pd.concat([df, file_df], ignore_index=True)
        return df
    except:
        None
    pass

def fetchConvertDataFiles(idx, path, df):
    for root, dirs, files in os.walk(path):
        for file in files:
            file_path = os.path.abspath(os.path.join(path, file))
            df = convertDataFile(idx, file_path, df)
            idx += 1
        for dir in dirs:
            idx += 1
            dir_path = os.path.abspath(os.path.join(path, dir))
            idx, df = fetchConvertDataFiles(idx, dir_path, df)

    return idx, df
```

```python
idx = 0
df = pd.DataFrame()
_, df = fetchConvertDataFiles(idx, data_path, df)
```
